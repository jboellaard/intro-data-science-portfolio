# Multivariate Analysis
### Clustering

```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
penguins.head()
```

```python
sns.pairplot(penguins)
plt.show()
```

Kijkend naar dit overzicht lijken er 2 duidelijke clusters te zijn voor body mass en flipper length. Verder lijkt het ook dat er 3 clusters zijn tussen bijvoorbeeld body mass en bill length en flipper length en bill length. 

```python
features = ['flipper_length_mm']
km = KMeans(n_clusters=2, random_state=43).fit(penguins[features])
```

```python
penguins['cluster'] = km.predict(penguins[features])
```

```python
sns.pairplot(penguins, hue="cluster")
plt.show()
```

```python
metrics.silhouette_score(penguins[features], km.labels_, metric='euclidean')
```

Na wat uitproberen blijkt dat twee clusters zoeken met flipper length de beste silhouette score geeft. 

```python
contingency_table = penguins.groupby(['sex','cluster']).size().unstack('sex', fill_value=0)
contingency_table.plot(kind='bar')
plt.show()
```

```python
contingency_table = penguins.groupby(['species','cluster']).size().unstack('species', fill_value=0)
contingency_table.plot(kind='bar')
plt.show()
```

```python
contingency_table = penguins.groupby(['island','cluster']).size().unstack('island', fill_value=0)
contingency_table.plot(kind='bar')
plt.show()
```

Van deze clusters zien we een scheiding tussen de species Gentoo en de andere species. Het verschil in flipper length tussen Adelie en Chinstrap net zoals Dream en Torgersen is dus blijkbaar niet zo groot. 

```python
sns.barplot(y="flipper_length_mm", x='species', data=penguins)
plt.show()
```

```python
sns.barplot(y="flipper_length_mm", x='island', data=penguins)
plt.show()
```

Het is het wel waard om nog te kijken of er een combinatie is van features die ofwel island ofwel species beter kan voorspellen met drie clusters.

```python
features_3 = ['flipper_length_mm', 'bill_length_mm']
km_3 = KMeans(n_clusters=3, random_state=43).fit(penguins[features_3])
```

```python
penguins['cluster_3'] = km_3.predict(penguins[features_3])
```

```python
metrics.silhouette_score(penguins[features_3], km_3.labels_, metric='euclidean')
```

Je krijg met alleen massa de hoogste silhouette score. Als je echter gaat kijken naar hoe het over species en eilanden is verdeeld, zie je dat een combinatie van flipper lenght en bill length zorgt dat Adelie beter gescheiden is van Chinstrap. 

```python
contingency_table = penguins.groupby(['cluster_3', 'species']).size().unstack('species', fill_value=0)
contingency_table.plot(kind='bar')
plt.show()
```

```python
contingency_table = penguins.groupby(['cluster_3','island']).size().unstack('island', fill_value=0)
contingency_table.plot(kind='bar')
plt.show()
```

```python
sns.pairplot(penguins.drop(['cluster'], axis=1), hue="species")
plt.show()
```
