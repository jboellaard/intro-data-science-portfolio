# Distributions
### Two numerical columns

```python
import pandas as pd
```

```python
data = pd.read_csv('star_dataset.csv')
data.head()
```

```python
data['Luminosity(L/Lo)'].plot(kind='hist', bins=10)
```

```python
data['Absolute magnitude(Mv)'].plot(kind='hist', bins=10, logx=True)
```

```python
data['Temperature (K)'].plot(kind='hist', bins=10)
```

```python
data['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

Er is voor deze dataset gekozen om een gelijke verdeling aan star types te selecteren (dwergen, main sequence en giants). Daarom is er een kolommen niet zomaar een normaalverdeling te vinden, er is een bias naar kleinere sterren (ook omdat de giants vele maten groter zijn). Je kan wel kijken naar rijen die gefilterd zijn op bijvoorbeeld kleur of type.

```python
color = 'Red'
data[(data['Star color'] == color)]['Temperature (K)'].plot(kind='hist', bins=10)
```

```python
data[(data['Star type'] == 'Red Dwarf')]['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

```python
data[(data['Star type'] == 'Brown Dwarf')]['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

```python
data[(data['Star type'] == 'Main Sequence')]['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

```python
data[(data['Star type'] == 'Super Giant')]['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

```python
data[(data['Star type'] == 'Hyper Giant')]['Radius(R/Ro)'].plot(kind='hist', bins=10)
```

Bij deze slices van data lijken er wel normal distributions aanwezig te zijn.
