# Bivariate analysis
### Pearson correlation and scatter plot

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('star_dataset.csv')
data.head()
```

```python
dataFrame = pd.DataFrame({'Temperature (K)':data['Temperature (K)'], 'Luminosity(L/Lo)':data['Luminosity(L/Lo)'], 'Radius(R/Ro)':data['Radius(R/Ro)'], 'Absolute magnitude(Mv)':data['Absolute magnitude(Mv)'] })
correlations = dataFrame.corr()
correlations.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

De hoogste correlatie die we hier zien is de absolute magnitude met de luminosity.

```python
data.plot(kind='scatter', x='Absolute magnitude(Mv)', y='Luminosity(L/Lo)')
```

Deze correlatie lijkt op een logaritmische schaal, en dat is logisch, want de formule voor Absolute Magnitude is:
$$ M_v = {4.74 - 2.5 * \log{_{10}}{L \over Lo}} $$

```python
data.plot(kind='scatter', x='Absolute magnitude(Mv)', y='Luminosity(L/Lo)', logy=True)
```

```python
data['L_to_M'] = 4.74 - 2.5*np.log10(data['Luminosity(L/Lo)'])
data.plot(kind='scatter', x='Absolute magnitude(Mv)', y='L_to_M')
```

```python
dataFrame_L_to_M = pd.DataFrame({'Temperature (K)':data['Temperature (K)'], 'Luminosity(L/Lo)':data['L_to_M'], 'Radius(R/Ro)':data['Radius(R/Ro)'], 'Absolute magnitude(Mv)':data['Absolute magnitude(Mv)'] })
correlations_L_to_M = dataFrame_L_to_M.corr()
correlations_L_to_M.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

De radius heeft ook een correlatie met beide absolute magnitude en de luminosity, wat ook logisch is, want bij grotere sterren verwacht je dat de totale hoeveelheid energie die per seconde wordt uitgestraald ook groter is.

```python
data.plot(kind='scatter', x='Radius(R/Ro)', y='Absolute magnitude(Mv)', logx=True)
```

De kleinste correlatie (bijna 0) is tussen Radius en Temperature. We verwachten hier ook geen correlatie omdat sterren van verschillende types (Main Sequence, Giants) dezelfde temperatuur kunnen hebben. 

```python
data.plot(kind='scatter', x='Radius(R/Ro)', y='Temperature (K)')
```
