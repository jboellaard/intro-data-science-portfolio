```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('dataset.csv')
data.head()
```

```python
category = "Star type"
data.groupby(category).mean()
```

```python
sns.barplot(y="Temperature (K)", x=category, data=data)
```

```python
sns.barplot(y="Luminosity(L/Lo)", x=category, data=data)
```

```python
sns.barplot(y="Radius(R/Ro)", x=category, data=data)
```

```python
sns.barplot(y="Absolute magnitude(Mv)", x=category, data=data)
```
