```python
import pandas as pd
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
dataFrame = pd.DataFrame({'Temperature (K)':data['Temperature (K)'], 'Luminosity(L/Lo)':data['Luminosity(L/Lo)'], 'Radius(R/Ro)':data['Radius(R/Ro)'], 'Absolute magnitude(Mv)':data['Absolute magnitude(Mv)'] })
```

```python
correlations = dataFrame.corr()
correlations.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
data.plot(kind='scatter', x='Absolute magnitude(Mv)', y='Luminosity(L/Lo)')
```

```python
data.plot(kind='scatter', x='Radius(R/Ro)', y='Absolute magnitude(Mv)')
```

```python
data.plot(kind='scatter', x='Radius(R/Ro)', y='Luminosity(L/Lo)')
```
