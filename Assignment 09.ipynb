```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
dataFrame = pd.DataFrame({'bill_length_mm':penguins['bill_length_mm'], 'bill_depth_mm':penguins['bill_depth_mm'], 'flipper_length_mm':penguins['flipper_length_mm'], 'body_mass_g':penguins['body_mass_g'] })
```

```python
correlations = dataFrame.corr()
correlations.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
correlations.plot(figsize=(15,8))
```

```python
penguins.plot(kind='scatter', x='flipper_length_mm', y='body_mass_g')
```

```python
penguins.plot(kind='scatter', x='flipper_length_mm', y='bill_depth_mm')
```
